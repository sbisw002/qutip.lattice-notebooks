## Spinless/ single-spin fermions in 1d
Defining an instance of the Lattice1d class requires formatting the second Quantized Hamiltonian in a unitcell based structure with nearest neighbor coupling only. Howewver, the functionality is limited to single particle physics only in Lattice1d class methods.

\begin{eqnarray}
H = -J\sum_{i} \sigma^i_z \sigma^{i+1}_z + h\sum_{i} \sigma^{i+1}_x
\end{eqnarray}

where $\sigma^i$ is the ith spin operator, J is the hopping, h is the magnetic field.

```python
from qutip import *
from qutip_lattice import *

import matplotlib.pyplot as plt
import numpy as np
```

# Instance of the 1D TFIM model

```python
Ising_Lattice1d = Lattice1d_TFIM(num_sites=4, boundary="periodic", J=1, h=1)
```

```python
[Hamiltonian_NS, basisStates_NS] = Ising_Lattice1d.Hamiltonian(filling=None, kval=None)
```

```python
UsF = Ising_Lattice1d.NoSym_DiagTrans()
```

```python
plt.spy(np.abs(Hamiltonian_NS.full()))
```

```python

###############################################################################
Hamiltonian_k = UsF * Hamiltonian_NS * UsF.dag()
nB = np.shape(Hamiltonian_k)[0]
Hamiltonian_k = Hamiltonian_k.full()
for i in range(nB):
    for j in range(nB):
        if np.abs(Hamiltonian_k[i,j]) < 1e-8:
            Hamiltonian_k[i, j] = 0
###############################################################################
```

```python
plt.spy(np.abs(Hamiltonian_k))
```

```python
[Hamiltonian_k0, basisStates_k0] = Ising_Lattice1d.Hamiltonian(filling=None, kval=0)
[Hamiltonian_k1, basisStates_k1] = Ising_Lattice1d.Hamiltonian(filling=None, kval=1)
[Hamiltonian_k2, basisStates_k2] = Ising_Lattice1d.Hamiltonian(filling=None, kval=2)
[Hamiltonian_k3, basisStates_k3] = Ising_Lattice1d.Hamiltonian(filling=None, kval=3)
```

```python
[Hamiltonian_N, basisStates_N] = Ising_Lattice1d.Hamiltonian(filling=2, kval=None)
UsFN = Ising_Lattice1d.nums_DiagTrans(filling=2)
```

```python
Hamiltonian_Nk = UsFN * Hamiltonian_N * UsFN.dag()
nB = np.shape(Hamiltonian_Nk)[0]
Hamiltonian_Nk = Hamiltonian_Nk.full()
for i in range(nB):
    for j in range(nB):
        if np.abs(Hamiltonian_Nk[i,j]) < 1e-8:
            Hamiltonian_Nk[i, j] = 0
```
